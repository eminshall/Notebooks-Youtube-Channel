```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels import regression
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import yfinance as yf
from datetime import datetime, date
import datetime
```

```python
ETFs = ['XLV','VHT','IBB','XBI','IHI','IXJ','IYH','FHLC']
healthcare = ['LLY','UNH','JNJ','ABBV','MRK','ABT','TMO','PFE']
index = '^GSPC'

end = date.today()
start = end - datetime.timedelta(days = 3*365)

def get_returns(portfolio, start, end):
  #download data
  data = yf.download(portfolio, start, end)['Close']

  #Calculate Returns
  portfolio_returns = (data
                       .pct_change()
                       .dropna()
                       .sum(axis=1))

  #Return portfolio Return Values
  return portfolio_returns

health_returns = get_returns(healthcare, start, end)
index_returns = get_returns(index, start, end)

health_returns.name = 'Healthcare'
index_returns.name = 'S&P 500'
```

```python
health_returns.plot()
index_returns.plot()
plt.ylabel('Daily Returns')
plt.title('Healthcare Returns vs. S&P 500 Returns')
plt.legend()
```

```python
xlv_returns = get_returns(ETFs[0], start, end)
xlv_returns.name = 'XLV'
vht_returns = get_returns(ETFs[1], start, end)
vht_returns.name = 'VHT'
ibb_returns = get_returns(ETFs[2], start, end)
ibb_returns.name = 'IBB'
xbi_returns = get_returns(ETFs[3], start, end)
xbi_returns.name = 'XBI'
ihi_returns = get_returns(ETFs[4], start, end)
ihi_returns.name = 'IHI'
ixj_returns = get_returns(ETFs[5], start, end)
ixj_returns.name = 'IXJ'
iyh_returns = get_returns(ETFs[6], start, end)
iyh_returns.name = 'IYH'
fhlc_returns = get_returns(ETFs[7], start, end)
fhlc_returns.name = 'FHLC'
etf_returns = [xlv_returns, vht_returns, ibb_returns, xbi_returns, ihi_returns, ixj_returns, iyh_returns, fhlc_returns]
```

```python
def information_ratio(portfolio_returns, bench_returns):
  active_return = portfolio_returns - bench_returns
  tracking_error = active_return.std()
  return active_return.mean() / tracking_error
```

```python
for etf in etf_returns:
  print(f'{etf.name} ETF Information Ratio: {information_ratio(etf, index_returns)}')

print(f'Healthcare Sector Information Ratio: {information_ratio(health_returns, index_returns)}')
```
